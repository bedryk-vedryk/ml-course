1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10,10)
def f(x):
  return 10*x - 5
def df(f, x, h=1e-6):
  return (f(x+h) - f(x))/h

y = f(x)
df_values =  df(f, x)
print(f"похідна функції 10х - 5: {round(df_values[0],2)}")
plt.figure(figsize=(10,10))
plt.plot(x, y, label='f(x) = 10*x - 5', color='blue')
plt.plot(x, df_values, label="f'(x) = 10", color='red')
plt.title('Графік функції та її похідної')
plt.xlabel('години витрачені на навчання')
plt.ylabel('оцінка')
plt.legend()
plt.grid(True, which='major')

plt.show()
```

Оскільки графік функції звичайна лінійна функція, залежність годин присвячених навчанню та оцінки, отриманої в результаті, також лінійна: чим більше годин відведено на навчання, тим вища оцінка.


2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

x = sp.symbols('x')

f_sin = sp.sin(x)
d_f_sin = sp.diff(f_sin, x)

f_ln = sp.ln(x)
d_f_ln = sp.diff(f_ln, x)

print(f"Похідна функції sin(x): {d_f_sin}")
print(f"Похідна функції ln(x): {d_f_ln}")

f_sin_np = sp.lambdify(x, f_sin, 'numpy')
d_f_sin_np = sp.lambdify(x, d_f_sin, 'numpy')
f_ln_np = sp.lambdify(x, f_ln, 'numpy')
d_f_ln_np = sp.lambdify(x, d_f_ln, 'numpy')

x_vals = np.linspace(0.1, 10, 100)
# Use the numerical functions to generate y-values
plt.plot(x_vals, f_sin_np(x_vals), label="$f(x)=sin(x)$", color='blue')
plt.plot(x_vals, d_f_sin_np(x_vals), linestyle='--', label="$f'(x)=cos(x)$", color='red')
plt.plot(x_vals, f_ln_np(x_vals), label="$f(x)=ln(x)$", color='green')
plt.plot(x_vals, d_f_ln_np(x_vals), linestyle='--', label="$f'(x)= \\frac{1}{x}$", color='orange')
plt.legend()
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols('x')
f = x**2 - 5*x + 1
df = sp.diff(f,x)

print(f"Похідна функції: {df}")

f_np = sp.lambdify(x, f, 'numpy')
df_np = sp.lambdify(x, df, 'numpy')
x = np.linspace(0,100,100)
plt.plot(x, f_np(x), label="$ f(x)=x^2 - 5x + 1 $", color='blue')
plt.grid(True, which='major')
plt.plot(x, df_np(x), linestyle='--', label="$ f'(x)=2x - 5 $", color='red')
plt.title('Графік функції та її похідної')
plt.xlabel('кількість вироблених одиниць')
plt.ylabel('загальні витрати виробництва')
plt.legend()
plt.show()
```

Похідна данної функції обчислює додаткові витрати ще однієї одиниці продукції (marginal cost)
