
### Завдання 3: Множення матриць
1. Задані дві матриці `A` і `B`розмірностями 3x2 та 2х3. Виконайте їхнє множення, щоб отримати матрицю 2x2. Виведіть результуючу матрицю `result_matrix` на екран.

2. Заданий вектор `C` розмірністю 2х1. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите `result_matrix` на вектор `C` і який результат буде. А потім виконайте множення і виведіть результат на екран.

3. Задана матриця `D` розмірністю 2х2. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите `result_matrix` на матрицю `D` і який результат буде. А потім виконайте множення і виведіть результат на екран.

```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

C = np.array([[0], [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

```python
# для того, щоб отримати матрицю розміром 2х2, потрібно помножити матрицю В(2х3) на матрицю А(3х2)
result_matrix = B @ A
print(f'результат множення матриць В та А: \n{result_matrix}')
# оскільки розмір матриці добутку 2х2, а розмір матриці С 2х1, результат їхнього множення буде матриця розміром 2х1
result_matrix @ C
print(f'результат множення матриць BA та С: \n{result_matrix @ C}')
# оскільки розмір матриці добутку 2х2, а розмір матриці Д теж 2х2, резульат їхнього множення буде теж матриця 2х2
result_matrix @ D
print(f'результат множення матриць BA та D: \n{result_matrix @ D}')
# важливо розуміти, що попри однаковий розмір матриць, перестановка матриць в множенні змінює результат їхнього добутку!
D @ result_matrix
print(f'результат множення матриць D та ВА: \n{D@result_matrix}')
```
